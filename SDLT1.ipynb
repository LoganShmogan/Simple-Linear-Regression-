# SDLT 01 - Simple Linear Regression

#### Steps 1 & 2

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import Dataset
dataset = pd.read_csv("AgeCholesterol.csv")
```

#### Step 3

```python
# Give dataset values
x = dataset.iloc[:,0].values
y = dataset.iloc[:,1].values
```

#### Steps 4 & 5

```python
# Plotting data
plt.ylabel("Age of Person")
plt.xlabel("Cholesterol level")
plt.title("Age of Person VS Cholesterol")
plt.scatter(x, y)
plt.show()
```

#### Steps 6 & 7

```python
model = np.polyfit(x, y, 1)
print(model)

# y = ax - b
# y = 3.74419576e-02 * x - 4.51457294e+01
```

#### Step 8

```python
# Prediction
predict = np.poly1d(model)
print(predict)

# Graph
plt.ylabel("Age of Person (AOP)")
plt.xlabel("Cholesterol leveli (CHL)")
plt.title("AOP vs CHL with Linear Model line")
plt.scatter(x, y)
plt.plot(x, np.polyval(model, x), 'r--')
plt.show()
```

#### Step 9

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(x, y)
reg.coef_
reg.intercept_


#x_lin_reg = range(0, 650)
#y_lin_reg = predict(x_lin_reg)
#plt.scatter(x, y)
#plt.plot(x_lin_reg, y_lin_reg, c = 'r')
```
